```python
import numpy as np
import matplotlib.pyplot as plt 
 
  
# creating the dataset
data = {'WWII': 101, 'Olympics': 54, 'Olympic': 53, 'WWI': 38, 
        'the Civil War': 37, 'World Series': 32, 'World War II': 32, 
        'Civil War': 28, 'World War I': 24, 'Super Bowl': 24}
keys = list(data.keys())
values = list(data.values())
  
fig = plt.figure(figsize = (20, 5))
 
# creating the bar plot
max_y_lim = 250
min_y_lim = 0
plt.ylim(min_y_lim, max_y_lim)
plt.bar(keys, values, color ='lightblue', 
        width = 0.4)
 
plt.xlabel("Event")
plt.ylabel("Frequency")
plt.title("Top 10 Most Common Event Entities from 2006 - 2012 Jeopardy Questions")
plt.show()
```
